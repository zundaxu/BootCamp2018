To solve the value function, we could use a "guess and verify" approach.  Plugging $K_{t+1}=Ae^{z_t}K_t^{\alpha}$ into both sides of the Euler Equation, we have:

$$\begin{align*}
\frac{1}{(1-A)e^{z_t}K_t^{\alpha}} = \beta E\{\frac{\alpha e^{z_{t+1}}(Ae^{z_t}K_t^{\alpha})^{\alpha-1}}{e^{z_{t+1}}(Ae^{z_t}K_t^{\alpha})^{\alpha}-Ae^{z_{t+1}}(Ae^{z_t}K_t^{\alpha})^{\alpha}}\}
\end{align*}$$

Simplifying the RHS of the equation, we get:

$$A = \alpha\beta$$

Then we can solve the value of $k$ by plug $\bar{r}$ into the firm's FOC for capital $$\bar{r} = \alpha\bar{k}^{\alpha-1}$$

** Numerical Solution:**

```python
def get_cbar(w, r, k, l):
    c_bar = w * l + (r - delta) * k
    return c_bar

def get_rbar(k, l):
    r_bar = alpha * (l/k) ** (1 - alpha) * (np.exp(Z_bar)) ** (1 - alpha)
    return r_bar

def get_wbar(k, l):
    w_bar = (1 - alpha) * (k/l) ** alpha * (np.exp(Z_bar)) ** (1 - alpha)
    return w_bar
    
def euler_equation(soln, *args):
    gamma, epsilon, beta, alpha, a, delta, tau, Z_bar = args
    w = get_wbar(soln, 1)
    r = get_rbar(soln, 1)
    c = get_cbar(w, r, soln, 1)
    errors = c ** (-gamma) - beta * (c ** (-gamma)* ((1 - tau) * (r - delta) + 1))
    return errors

def get_Tbar(w, r, k, l):
    T_bar = tau * (w * l + (r - delta) * k)
    return T_bar

def get_ibar(k):
    i_bar = delta * k
    return i_bar

def get_ybar(k, l):
    y_bar = k ** alpha * l ** (1 - alpha)
    return y_bar
```

```python
args1 = (gamma, epsilon, beta, alpha, a, delta, tau, Z_bar)
result1 = opt.fsolve(euler_equation, 1, args = (args1))
w1 = get_wbar(result1[0], 1)
r1 = get_rbar(result1[0], 1)
T1 = get_Tbar(w1, r1, result1[0], 1)
i1 = get_ibar(result1[0])
y1 = get_ybar(result1[0], 1)
print("steady-state capital is: ", result1[0])
print("steady-state wage is: ", w1)
print("steady-state interest rate is: ", r1)
print("steady-state consumption is: ", get_cbar(w1, r1, result1[0], 1))
print("steady-state government transfer is: ",T1)
print("steady-state investment is: ",i1)
print("steady-state output is: ",y1)
```

From the results shown above, we note that the algebraic solution and the numerical solution are the same.

#### Problem 1:

To linearize, we need to write a gamma function in terms of the state variables, based on the Euler Equation.

The Euler equation for the Brock-Mirman Model is: 

$$\frac{1}{e^{z_t}K_t^\alpha - K_{t+1}} = \beta E_t\{\frac{\alpha e^{z_{t+1}}K_{t+1}^{\alpha-1}}{e^{z_{t+1}}K_{t+1}^{\alpha}-K_{t+2}}\}$$

Rewritting the associated Euler equation, we get:

$$\Gamma(K_{t+2}, K_{t+1}, K_{t}, z_{t+1}, z_{t}) = \frac{1}{e^{z_t}K_t^\alpha - K_{t+1}} = \beta E_t\{\frac{\alpha e^{z_{t+1}}K_{t+1}^{\alpha-1}}{e^{z_{t+1}}K_{t+1}^{\alpha}-K_{t+2}}\}$$

```python
# Therefore we could define a function to get the value of the rewritten associated euler equation shown above, 
# And this function can also be used to check the steady solution since the value equals to zero at the steady state.
def associated_euler(theta0, params):
    '''
    Inputs:
        theta0: a vector containng (K_t2, K_t1, K_t, Z_t1, Z_t)
        params: list of parameter values
    
    Output:
        Euler: an associated Euler equation rewritten so that they are zero at the
               steady state values of X, Y & Z.  
    '''
    (K_t2, K_t1, K_t, Z_t1, Z_t) = theta0
    Euler = 1/(np.exp(Z_t) * K_t ** alpha - K_t1) - beta * ((alpha * np.exp(Z_t1) * K_t1 ** (alpha - 1))/ 
                                             (np.exp(Z_t1) * K_t1 ** alpha - K_t2))
                                                                            
    return Euler
```

#### Problem 4

By letting the $k_{t-1}$ be the state variables, we can solve for the steady-state values using the LinApp functions.

```python
def get_c(w, r, kp, k, l):
    c = w * l + (r - delta) * k + k - kp
    return c

def get_r(k, l, z):
    r = alpha * (l/k) ** (1 - alpha) * (np.exp(z)) ** (1 - alpha)
    return r

def get_w(k, l, z):
    w = (1 - alpha) * (k/l) ** alpha * (np.exp(z)) ** (1 - alpha)
    return w
```

```python
def euler_equation(theta0, params):
    (X_t2, X_t1, X_t, Y_t1, Y_t, Z_t1, Z_t) = theta0
    w_t = get_w(X_t, Y_t, Z_t)
    w_t1 = get_w(X_t1, Y_t1, Z_t1)
    r_t = get_r(X_t, Y_t, Z_t)
    r_t1 = get_r(X_t1, Y_t1, Z_t1)
    c_t = get_c(w_t, r_t, X_t1, X_t, Y_t)
    c_t1 = get_c(w_t1, r_t1, X_t2, X_t1, Y_t1)

    E1 = a * (1 - Y_t) ** (-epsilon) - (c_t ** (-gamma) * w_t * (1 - tau)) 
    E2 = c_t ** (-gamma)  - beta * (c_t1 ** (-gamma)* ((1 - tau) * (r_t1 - delta) + 1))
    return np.array([E1, E2])
```

```python
def get_c(w, r, kp, k, l):
    c = w * l + (r - delta) * k + k - kp
    return c

def get_r(k, l, z):
    r = alpha * (l/k) ** (1 - alpha) * (np.exp(z)) ** (1 - alpha)
    return r

def get_w(k, l, z):
    w = (1 - alpha) * (k/l) ** alpha * (np.exp(z)) ** (1 - alpha)
    return w
```

```python
def euler(theta0, params):
    (X_t2, Y_t2, X_t1, Y_t1, X_t, Y_t, Z_t1, Z_t) = theta0
    w_t = get_w(X_t, Y_t1, Z_t)
    w_t1 = get_w(X_t1, Y_t2, Z_t1)
    r_t = get_r(X_t, Y_t1, Z_t)
    r_t1 = get_r(X_t1, Y_t2, Z_t1)
    c_t = get_c(w_t, r_t, X_t1, X_t, Y_t1)
    c_t1 = get_c(w_t1, r_t1, X_t2, X_t1, Y_t2)
    
    E1 = a * (1 - Y_t1) ** (-xi)  - (c_t ** (-gamma) * w_t * (1 - tau))
    E2 = c_t ** (-gamma) - ( beta * (c_t1 ** (-gamma)* ((1 - tau) * (r_t1 - delta) + 1))) 
    
    return np.array([E1, E2])
```

```python
# Set Parameters
nvec = np.array([1.0,1.0,0.2]) # labor supply
yrs_live = 60
s = 3
beta = 0.442
sigma = 3
alpha = 0.35
delta_annual = 0.05
delta = 1- ((1 - delta_annual) ** (yrs_live/s))
params = (nvec, beta, sigma, alpha)
nx = 2
ny = 0
nz = 1
zbar = np.array([0.])
guessk = np.array([.02, .04])
```

```python
# get aggregate variables
def get_L(nvec): 
    L = nvec.sum()
    return L

def get_K(kvec): 
    K = kvec.sum()
    return K
#use characteristic eqs to get w and r
def get_w(K, L, z): 
    w = (1 - alpha) * np.exp(z) * ((K / L) ** alpha)
    return w

def get_r(K, L, z): 
    r = alpha * np.exp(z) * ((L / K) ** (1 - alpha)) - delta
    return r
```

```python
def Model_olg(theta0, params):
    (k2pp, k3pp, k2p, k3p, k2, k3, zp, z) = theta0
    Kpp = get_K(np.array([k2pp, k3pp]))
    Kp = get_K(np.array([k2p, k3p]))
    K = get_K(np.array([k2, k3]))
    L = get_L(nvec)
    wp = get_w(Kp, L, zp)
    w = get_w(K, L, z)
    rp = get_r(Kp, L, zp)
    r = get_r(K, L, z)
    c1 = w * nvec[0] - k2p
    c2 = (1 + r) * k2 + w * nvec[1] - k3p
    c2p = (1 + rp) * k2p + wp * nvec[1] - k3pp
    c3p = (1 + rp) * k3p + wp * nvec[2]
    # errors
    E1 = c1 ** (-sigma) - beta * (1 + rp) * c2p ** (-sigma)
    E2 = c2 ** (-sigma) - beta * (1 + rp) * c3p ** (-sigma)
    
    return np.array([E1, E2])
```

```python
k2bar, k3bar = LinApp_FindSS(Model_olg, params, guessk, zbar, nx, ny)
theta0 = np.array([k2bar, k3bar, k2bar, k3bar, k2bar, k3bar, 0., 0.])
```

```python
def error_n(w, *args):
    alpha, z, k, b, t, h = args
    nd = ((1 - alpha) * z/w) ** (1/alpha) * k
    pi = z * k ** alpha* nd ** (1 - alpha) - w * nd
    ns = h - b/(w * (1 + b)) * (w * h + pi -t)
    return nd - ns
```

```python
args = (alpha, z, k, b, t, h)
w = opt.fsolve(error_n, 0.8, args = (args))[0]
print(w)
```

```python
def first_deriv(f, x0, h=1e-2):
    return (f(x0+h)-f(x0-h))/(2*h)

def second_deriv(f, x0, h=1e-2):
    return (f(x0+h)+f(x0-h)-2*f(x0))/(h**2)

def get_wss(k1, w_init):
    args = (alpha, z, k1, b, t, h)
    w = opt.fsolve(error_n, w_init, args = (args))[0]
    return w

def first_approx(f, x0, x):
    fp = first_deriv(f, x0)
    return f(x0) + fp * (x - x0)
    
def second_approx(f, x0, x):
    fp = first_deriv(f, x0)
    fpp = second_deriv(f, x0)
    return f(x0) + fp * (x - x0) + (1/2) * fpp * (x - x0) ** 2

f = lambda x: get_wss(x, 0.8)
print(first_approx(f, 5, 5.1))
print(second_approx(f, 5, 5.1))
```

```python
e = 0.001

F = (Lin_SS(k2=kbar+e) - Lin_SS(k2=kbar-e))/(2 * e)
G = (Lin_SS(k1=kbar+e) - Lin_SS(k1=kbar-e))/(2 * e)
H = (Lin_SS(k=kbar+e) - Lin_SS(k=kbar-e))/(2 * e)   
L = (Lin_SS(z1=zbar+e) - Lin_SS(z1=zbar-e))/(2 * e)
M = (Lin_SS(z=zbar+e) - Lin_SS(z=zbar-e))/(2 * e)
N = 0.95
Gamma_A = np.array([F, G, H, L, M])

Hx = (-G - np.sqrt(G ** 2 - 4 * F * H))/(2 * F)
Hz = -(L * N + M)/(F * N + F * Hx + G)
print("H_x is:",Hx)
print("H_z is:",Hz)
```
